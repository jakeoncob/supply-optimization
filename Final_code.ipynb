```python
def clean(txt):
    import re, unidecode
    txt = unidecode.unidecode(str(txt)).upper()
    txt = re.sub(r"T[P]?\.?\s+", "", txt)   
    txt = txt.replace("-", " ")
    return re.sub(r"\s+", " ", txt).strip()
list_of_SKU_Province = [clean(p) for p in list_of_SKU_Province]

```

```python
def increase_demand(d,pct):
    return {
        p:{
            i: int(round(v * (1 + pct)))
            for i,v in list_of_SKU_ID.items()

        }
        for p, list_of_SKU_ID in d.items()
    }
```

```python
import random

def increase_demand_random(demand_dict: dict, pct: float = 0.15, seed=None):

    rng  = random.Random(seed)         
    newd = {}

    for prov, sku_map in demand_dict.items():
        
        bumped = {sku: int(round(qty * (1 + pct))) for sku, qty in sku_map.items()}

        base_total    = sum(sku_map.values())
        target_total  = int(round(base_total * (1 + pct)))
        current_total = sum(bumped.values())
        shortfall     = target_total - current_total   

        
        if shortfall > 0 and bumped:
            sku_ids = list(bumped.keys())
            for _ in range(shortfall):
                bumped[rng.choice(sku_ids)] += 1

        newd[prov] = bumped

    return newd
```

```python
def solve_transport(supply: dict,
                    demand: dict,
                    cost: dict,
                    surge_limit: float = 1, #anything past 1 is now the percent a farm is allowed to send past its current supply (must be increased greatly if demand is increased)
                    penalty_factor: float=2.0, # the penalty for using surge units so the optimization knows they arent preferable
                    integer: bool = True):
    prob = pulp.LpProblem("Supply_Matrix_with_Surge",pulp.LpMinimize)
    cat = "Integer" if integer else "Continous"
    
    ship_norm = pulp.LpVariable.dicts("ShipNorm",
                                      (unique_farm_ids,list_of_SKU_Province,list_of_SKU_ID),
                                      lowBound=0,cat=cat)
    ship_surge = pulp.LpVariable.dicts("ShipSurge",
                                      (unique_farm_ids,list_of_SKU_Province,list_of_SKU_ID),
                                      lowBound=0,cat=cat)
    for f in unique_farm_ids:
        for i in list_of_SKU_ID:
            base_cap = supply.get(f,{}).get(i,0)
            prob += (pulp.lpSum(ship_norm[f][p][i] for p in list_of_SKU_Province) <= base_cap)
            prob += (pulp.lpSum(ship_norm[f][p][i] + ship_surge[f][p][i] for p in list_of_SKU_Province) <= surge_limit * base_cap)
    for p in list_of_SKU_Province:
        for i in list_of_SKU_ID:
            need = demand.get(p, {}).get(i,0)
            prob += (pulp.lpSum(ship_norm[f][p][i] + ship_surge[f][p][i] for f in unique_farm_ids) == need)
    
    def lane_cost(farm, prov, bigM=1e6):
        return cost.get(farm, {}).get(prov, bigM)
    
    prob += (
        pulp.lpSum(lane_cost(f, prov) * ship_norm[f][prov][sku]
                for f in unique_farm_ids
                for prov in list_of_SKU_Province
                for sku in list_of_SKU_ID)
        +
        penalty_factor *
        pulp.lpSum(lane_cost(f, prov) * ship_surge[f][prov][sku]
                for f in unique_farm_ids
                for prov in list_of_SKU_Province
                for sku in list_of_SKU_ID)
    )
    status = prob.solve(pulp.PULP_CBC_CMD(msg=False))

    norm_out = {f: {p: {i: ship_norm[f][p][i].value() for i in list_of_SKU_ID} for p in list_of_SKU_Province} for f in unique_farm_ids}

    surge_out = {f: {p: {i: ship_surge[f][p][i].value() for i in list_of_SKU_ID} for p in list_of_SKU_Province} for f in unique_farm_ids}
    



    return pulp.LpStatus[status], pulp.value(prob.objective),norm_out,surge_out
```

```python
print("No Change in demand")
status, obj, norm, surge = solve_transport(dictionary_of_farm_supply,dictionary_of_SKU_Province_supply,distance_farm_to_province)
print(f"Status: {status} Total Cost: {obj}")
for f in unique_farm_ids:
    for p in list_of_SKU_Province:
        for i in list_of_SKU_ID:
            q = norm[f][p][i] + surge[f][p][i]
            if q:
                note = f"(includes. {surge[f][p][i]} surge)" if surge[f][p][i] else ""
                print(f"{f} -> {p} [{i}]: {int(q)}{note}")
print()
```
